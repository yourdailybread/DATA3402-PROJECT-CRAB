```python
df.drop('id', axis=1, inplace=True)
df.Sex = df.Sex.map({'F': 0, 'M': 1, 'I': 2})
```

```python
df = df.rename(columns={'Shucked Weight': 'shucked_weight', 'Viscera Weight': 'viscera_weight', 'Shell Weight': 'shell_weight'})
df.columns
```

```python
plt.figure(figsize=(10, 6))
mask = np.triu(np.ones_like(df.corr(), dtype=bool))
sns.heatmap(df.corr(), mask=mask, annot=True, cmap='coolwarm',
                linewidths=0.2, annot_kws={"size": 7}, rasterized=True)
plt.xticks(rotation=55)
plt.show()
```
